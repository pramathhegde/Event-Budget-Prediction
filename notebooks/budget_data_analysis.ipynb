```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/prama/Downloads/historical_event_budgets.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["approved_budget"], bins=10, kde=True)
plt.title("Distribution of Approved Event Budgets")
plt.xlabel("Budget (â‚¹)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="participants", y="approved_budget", data=df)
plt.title("Participants vs Approved Budget")
plt.show()
```

```python
avg_budget = df.groupby("event_type")["approved_budget"].mean()

avg_budget.plot(kind="bar", figsize=(6,4), title="Average Budget by Event Type")
plt.ylabel("Average Budget (â‚¹)")
plt.show()
```

```python
dept_budget = df.groupby("department")["approved_budget"].mean()

dept_budget.plot(kind="bar", figsize=(6,4), title="Average Budget by Department")
plt.ylabel("Average Budget (â‚¹)")
plt.show()
```
